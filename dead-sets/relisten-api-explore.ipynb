# Download Dead Sets

[API](https://api.relisten.net/api-docs/index.html)

## Data we want

* CSV with every song and show
* JSON nested similarly to Relisten
* Raw JSON objects from each year


```python
import json
import requests
import urllib

api_url = "https://api.relisten.net/api/v2/"
slug = "grateful-dead"
```

```python
# Get artists
req_url = urllib.parse.urljoin(api_url, "artists")
r = requests.get (req_url)
print(r)
print(r.url)
#print(r.content)
#print(r.text)

parsed = json.loads(r.text)[0]
print(json.dumps(parsed, indent=4, sort_keys=True))
```

```python
# Get years
req_url = urllib.parse.urljoin(api_url, "artists/%s/years"%slug)
r = requests.get (req_url)
print(r)
print(r.url)
#print(r.content)
#print(r.text)

parsed = json.loads(r.text)
print(len(parsed))
parsed = parsed[0]
print(json.dumps(parsed, indent=4, sort_keys=True))
```

```python
# Get year
req_url = urllib.parse.urljoin(api_url, "artists/%s/years/%d"%(slug, 1965))
r = requests.get (req_url)
print(r)
print(r.url)
#print(r.content)
#print(r.text)

parsed = json.loads(r.text)
print(len(parsed))
#parsed = parsed[0]
print(json.dumps(parsed, indent=4, sort_keys=True))
```

```python
# Get show
req_url = urllib.parse.urljoin(api_url, "artists/%s/years/%d/%s"%(slug, 1969, "1969-01-17"))
r = requests.get (req_url)
print(r)
print(r.url)
#print(r.content)
#print(r.text)

parsed = json.loads(r.text)
print(len(parsed))
#parsed = parsed[0]
print(json.dumps(parsed, indent=4, sort_keys=True))
```

```python
# Get songs
req_url = urllib.parse.urljoin(api_url, "artists/%s/songs"%slug)
r = requests.get (req_url)
print(r)
print(r.url)
#print(r.content)
#print(r.text)

parsed = json.loads(r.text)
print(len(parsed))
#parsed = parsed[0]
print(json.dumps(parsed, indent=4, sort_keys=True))
```
